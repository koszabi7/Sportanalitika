```python
xg_df = pd.read_csv("xg_df_merged.csv")
min_results_df = pd.read_csv("merged_min_results.csv")
```

```python
xg_df = xg_df.drop(columns=['Unnamed: 18'], errors='ignore')
min_results_df = min_results_df.drop(columns=['Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26'], errors='ignore')
```

```python
merged_df = pd.merge(xg_df, min_results_df, on="id", how="inner")
```

```python
merged_df.head()
```

```python
merged_df = merged_df.drop(columns=['Unnamed: 0_x','Unnamed: 0_y','Notes', 'notes', 'Round', 'winner_OT', 'OT_method', 'Referee', 'Venue', 'Attendance'], errors='ignore')
```

```python
merged_df.columns
```

```python
merged_df = merged_df[merged_df["outcome"] == "Goal"]
```

```python
merged_df["Date"] = pd.to_datetime(merged_df["Date"])
```

```python
def get_season(date):
    year = date.year
    if date.month >= 8:  # August to December -> current year season
        return f"{year}-{year + 1}"
    else:  # January to May -> previous year season
        return f"{year - 1}-{year}"
```

```python
merged_df["season"] = merged_df["Date"].apply(get_season)
```

```python
seasonal_stats = merged_df.groupby(["player", "season"]).agg(
    total_xG=("xg", "sum"),
    total_goals=("outcome", lambda x: (x == "goal").sum()),  # Count only goals
).reset_index()
```
